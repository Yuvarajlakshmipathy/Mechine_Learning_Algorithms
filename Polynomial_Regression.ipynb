Polynomial Regession : Polynomial Regession is a type of regression analysis where the relationship
between the inpendant variable (input) and the dependent variable (output) is modeled as an nth degree polynomial. this allows for more flexbillity in fitting curves to the data, beyend simple linear relationships. a "shot" in this context likely refer to a data point or observation in the dataset being analyzed polynomial regression

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
sim_reg = LinearRegression()
sim_reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,sim_reg.predict(x),color="black")
plt.title("Gaming Analysis")
plt.xlabel("Steps")
plt.ylabel("Points")
plt.show()
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
x_poly = poly_reg.fit_transform(x)
sim_reg2 = LinearRegression()
sim_reg2.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,sim_reg2.predict(poly_reg.fit_transform(x)),color="green")
plt.title('Gaming Analysis (Polynomial Regression)')
plt.xlabel('Steps')
plt.ylabel('Points')
plt.show()
```
